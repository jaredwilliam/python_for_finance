In the CAPM setting, investors are risk-averse, prefer higher returns, and are willing to buy the optimal portfolio. 

The CAPM assumes the existence of a risk-free asset. An investment with zero risk. 

Why should do we assume the risk-free rate has a lower expected rate of return? In efficient markets, investors are only compensated for the added risk they are willing to bear. 

Investors will allocate their money between risk-free and the market portfolio. 

The line that connects the risk free rate and is tangent to the efficient frontier is called the capital market line. 

# Understanding and Calculating a Security's Beta

Beta helps us quantify the relationship between a security and the overall market portfolio. 

Safer stocks will earn less (and lose less) than the market portfolio. Riskier stocks will earn more (and lose more) than the market porfolio. 

Beta is calculated by taking the covariance of the stock divided by the variance of the market. It measures the market risk that cannot be avoided through diversification. The riskier the stock, the higher the beta. 

```python
from IPython.display import Image
Image(filename = 'beta.png')
```

Beta of 0 means there is no relationship. Beta < 1 is a defensive stock. Beta > 1 is an aggressive stock. 

```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb

tickers = ['PG', '^GSPC']
data = pd.DataFrame()
for t in tickers:
    data[t] = wb.DataReader(t, data_source='yahoo', start = '2012-1-1', end = '2016-12-31')['Adj Close']
```

```python
sec_returns = np.log(data / data.shift(1))
```

```python
cov = sec_returns.cov() * 250
cov
```

```python
cov_with_market = cov.iloc[0,1]
cov_with_market
```

```python
market_var = sec_returns['^GSPC'].var() * 250
market_var
```

```python
PG_beta = cov_with_market / market_var
PG_beta
```

# Introducing the Sharpe Ratio and Put it into Practice

Rational investors want to maximize their returns, but rational investors also want to minimize their risk and hence invest in less volatile securities. A rational investor considers both risk and return. Rational investors want to be able to compare stocks in terms of risk-return performance. 

```python
Image(filename = 'sharpe ratio.png')
```

If we increase a stock's expected rate of return, its Sharpe ratio will increase. If we increased the stock's standard deviation, the Sharpe ratio will decrease. The Sharpe ratio allows us to compare stocks or portfolios against other stocks or portfolios. 

```python
Sharpe = (PG_er - 0.025) / (sec_returns['PG'].std() * 250 ** 0.5)
Sharpe
```
