# Considering Both Risk and Return 

Have to consider both risk and return when making investment decisions. 

# Calculating a Security's Rate of Return

$rateOfReturn = \frac{endingPrice - beginningPrice}{beginningPrice}$

Logarithmic rate of return = log(ending price / beginning price)

Simple rate of return is preferable when dealing with multiple assets over the same timeframe and logarithmic rate of return is preferrable when you make calculations about a single asset over time. 

$annualReturn = [(dailyReturn + 1)^{365}]*100 -1$

This converts daily returns to yearly returns. Same logic applies to convert monthly or quarterly returns as well. 

```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
PG = wb.DataReader('PG', data_source='yahoo', start = '1995-1-1')
```

```python
PG.head()
```

```python
PG.tail()
```

## Simple Rate of Return

<center>$\frac{P_1 - P_0}{P_0} = \frac{P_1}{P_0}-1$</center>

```python
PG['simple_return'] = (PG['Adj Close'] / PG['Adj Close'].shift(1)) - 1
print(PG['simple_return'])
```

```python
PG['simple_return'].plot(figsize=(8,5))
```

```python
avg_returns_d = PG['simple_return'].mean() # average daily return
print(avg_returns_d)
```

```python
# Approximate return per year
avg_returns_a = PG['simple_return'].mean() * 250
print(avg_returns_a)
```

```python
print(str(round(avg_returns_a*100,3)), '%')
```

## Log Returns

<center>$ \ln(\frac{P_t}{P_{t-1}}) $</center>

```python
PG.head()
```

```python
PG['log_return'] = np.log(PG['Adj Close'] / PG['Adj Close'].shift(1))
print(PG['log_return'])
```

```python
PG['log_return'].plot(figsize=(8,5))
```

```python
log_return_d = PG['log_return'].mean()
log_return_d
```

```python
log_return_a = PG['log_return'].mean() * 250
log_return_a
```

```python
print(round(log_return_a*100,3), '%')
```

# Portfolio Securities and Calculate Rate of Return 

Calculating the rate of return of a portfolio that includes four stocks: Apple, IBM, GE, and Ford. 

rate of return for a security * weight in portfolio

```python
apple = {'rate': 0.318524, 'weight':  0.40}
ibm = {'rate': 0.153581, 'weight':  0.40}
ge = {'rate': 0.116717, 'weight':  0.15}
ford = {'rate': 0.126224, 'weight':  0.05}

rate_return = apple['rate']*apple['weight'] + ibm['rate']*ibm['weight'] + ge['rate']*ge['weight'] + ford['rate']*ford['weight']
rate_return
```

```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
tickers = ['PG', 'MSFT', 'F', 'GE']
mydata = pd.DataFrame()
for t in tickers:
    mydata[t] = wb.DataReader(t, data_source = 'yahoo', start = '1995-1-1')['Adj Close']
```

```python
mydata.info()
```

```python
mydata.head()
```

```python
mydata.tail()
```

## Normalization to 100

<center>$\frac{P_t}{P_0} * 100 $</center>

The idea behind this operation is to compare the value of the four different stocks as if they all started from the same value (100) (first plot). If we don't do this, we will see the values of the adjusted closing prices of the companies plotted directly, and they will start in different spots (second plot).

```python
mydata.iloc[0]
```

```python
(mydata / mydata.iloc[0] * 100).plot(figsize = (15,6))
```

```python
mydata.plot(figsize=(15,6))
```

## Calculating the Return of a Portfolio of Securities

```python
returns = (mydata / mydata.shift(1)) - 1
returns.head()
```

```python
weights = np.array([0.25, 0.25, 0.25, 0.25])
```

```python
np.dot(returns, weights)
```

```python
annual_returns = returns.mean() * 250
annual_returns
```

```python
np.dot(annual_returns, weights)
```

```python
pfolio_1 = str(round(np.dot(annual_returns, weights)*100,5)) + ' %'
print(pfolio_1)
```

# Calculating the Indices Rate of Return
