In the CAPM setting, investors are risk-averse, prefer higher returns, and are willing to buy the optimal portfolio. 

The CAPM assumes the existence of a risk-free asset. An investment with zero risk. 

Why should do we assume the risk-free rate has a lower expected rate of return? In efficient markets, investors are only compensated for the added risk they are willing to bear. 

Investors will allocate their money between risk-free and the market portfolio. 

The line that connects the risk free rate and is tangent to the efficient frontier is called the capital market line.
